```python
import csv
import matplotlib.pyplot as plt

f = open("all_temp.csv", 'r')
data = csv.reader(f)
header = next(data)

h =[]
l =[]

for row in data:
    if row[-1] != "" and row[-2] != "":
        if row[0].split("-")[1] == "01" and row[0].split("-")[2] == "30":
            h.append(float(row[4]))
            l.append(float(row[3]))
    
plt.title("my birthday")
plt.plot(h, label = "highest")
plt.plot(l, label = "lowest")
plt.legend()
plt.show()

```

```python
import csv
import matplotlib.pyplot as plt

f = open("seoul.csv", 'r')
data = csv.reader(f)
header = next(data)

J =[]
A =[]

for row in data:
    if row[-1] != "" and row[-2] != "":
        if row[0].split("-")[1] == "01":
            J.append(float(row[-1]))
        elif row[0].split("-")[1] == "08":
            A.append(float(row[-1]))

plt.title("highest temperature in 1,8")
plt.hist(J, bins = 100, color='b', label="Jan")
plt.hist(A, bins = 100, color='r', label="Aug")
plt.legend()
plt.show()
```
